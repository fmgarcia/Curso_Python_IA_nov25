```python
import pandas as pd
```

```python
df = pd.read_csv(r'..\ficheros\trx-10k.csv', sep=',', header='infer')
```

```python
df.describe() # generar estadísticas descriptivas del DataFrame, como la media, la desviación estándar, los valores mínimos y máximos, y los percentiles para las columnas numéricas
```

```python
# Cálculo de la media de la columna 'amount'
df['amount'].mean() # calcular la media de la columna 'amount' del DataFrame
```

```python
# Cálculo de la media de la columna 'amount' teniendo en cuenta solo las transacciones con status igual a 'success'
df_success = df[df['status'] == 'success'] # filtrar el DataFrame para incluir solo las filas donde la columna 'status' es igual a 'success'
df_success['amount'].mean() # calcular la media de la columna 'amount' del DataFrame filtrado
```

```python
df[df['status'] == 'success']['amount'].mean() # filtro con una condición y cálculo de la media en una sola línea
```

```python
df[(df['status'] == 'success') | (df['status'] == 'Success')]['amount'].mean() # filtro con múltiples condiciones y cálculo de la media en una sola línea
```
