```python
import csv
```

```python
# Ejemplo de lectura de un fichero CSV sin tratamiento de las cabeceras
with open(r'..\ficheros\datos.csv', newline='') as csvfile:
    filas = csv.reader(csvfile, delimiter=',', quotechar='"')
    for fila in filas:
        nombre, salario, revision = fila
        print(f"Nombre: {nombre}. Salario: {salario}. Próxima revisión: {revision}")
```

```python
# Ejemplo 1 de lectura de un fichero CSV con tratamiento de las cabeceras
with open(r'..\ficheros\datos.csv', newline='') as csvfile:
    filas = csv.reader(csvfile, delimiter=',', quotechar='"')
    next(filas)  # Saltar la fila de cabeceras
    for fila in filas:
        nombre, salario, revision = fila
        print(f"Nombre: {nombre}. Salario: {salario}. Próxima revisión: {revision}")
```
