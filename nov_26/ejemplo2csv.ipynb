```python
import csv
```

```python
# Ejemplo de lectura de un fichero CSV sin tratamiento de las cabeceras
with open(r'..\ficheros\datos.csv', newline='') as csvfile:
    filas = csv.reader(csvfile, delimiter=',', quotechar='"')
    for fila in filas:
        nombre, salario, revision = fila
        print(f"Nombre: {nombre}. Salario: {salario}. Próxima revisión: {revision}")
```

```python
# Ejemplo de lectura de un fichero CSV con tratamiento de las cabeceras
with open(r'..\ficheros\datos.csv', newline='') as csvfile:
    filas = csv.reader(csvfile, delimiter=',', quotechar='"')
    next(filas)  # Saltar la fila de cabeceras
    for fila in filas:
        nombre, salario, revision = fila
        print(f"Nombre: {nombre}. Salario: {salario}. Próxima revisión: {revision}")
```

```python
with open(r'..\ficheros\datos.csv', newline='') as csvfile:
    csvreader = csv.DictReader(csvfile, delimiter=',', quotechar='"')
    for fila in csvreader:
        print(f"Nombre: {fila['Nombre']}. Salario: {fila['Salario']}. Próxima revisión: {fila['Revisión']}")
```

```python
with open(r'..\ficheros\DSQA-full.csv', newline='') as csvfile:
    csvreader = csv.DictReader(csvfile, delimiter=',', quotechar='"')
    for fila in csvreader:
        print(f"Problema: {fila['problem'][:5]} Categoría: {fila['problem_category']}. Answer: {fila['answer']} Answer type: {fila['answer_type']}")
```
